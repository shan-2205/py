```python
'''

Hi there 👋

Now let’s learn something really powerful…
👉 How to create your own custom command in Claude Code.

Custom Slash Commands are those commands : which u can build for your workflows/or for your Projects as per need !


Lets continue with the same Sales analysis Project on which we were working on : 
Show [hierarchy]





'''
```

```python
'''

In order to create custom commans,  u have to 3 step procedure :
    lets understand !


✅ Step 1

    First, go inside our project folder…
    and create a new folder called : 👉 .claude

    Think of this as your toolbox 🧰
    where all Claude-related settings will live..


✅ Step 2

    Now, inside this .claude folder…
    create another folder called : 👉 commands

    You can think of this as a collection of tools…
    or simply, all your custom commands.



✅ Step 3
    Inside commands, create: 👉 explain.md
    This file will define what your command actually does.



    Inside explain.md  , put :
    Explain the sales_analysis.py code in a clear and simple way.

    - Break it step by step
    - Use beginner-friendly language
    - Explain purpose of each block
    - Avoid jargon

    If it's a full file, summarize it..




    So in simple words 

    .claude → your toolbox/lab
    commands → tools inside it
    .md file → actual custom command logic


    “and Claude Code automaticallye scans the folder .claude/commands/ to find out Custom commands defined by you”

    You don’t need to:

    install anything ❌
    restart manually (in most cases) ❌



If you simply type:

    👉 /explain

    Claude will automatically understand… and explain your entire project
    based on the instructions you defined.


So if your manager asks:
👉 “Explain your project…”

You won’t panic 😄

You’ll just run:

👉 /explain

    That’s the power of custom commands in Claude Code 🚀

'''
```

# 4.. Sessions !

```python
'''

Hi there , since we understood what commands are & how to define custom commands to automate your Workflows !

now lets understand one of the most IMPORTANT concepts in Claude Code which is : Sessions 

    So what is a session?
    👉 A session = One conversation with Claude
        Like here in this Project , [show vscode] , this entire Conversation that u have done to Claude
        to call custom command is 1 session



    So when u create a new terminal for Claude , 

    A session starts.
    & then you do all the conversation with Claude Code & When you exit that terminal , 👉 Session ends..
    
        So this entire in between conversation is actually one session !
        
        So everything in between…

        👉 your prompts
        👉 your code
        👉 your questions

All of that = one session
        

But here’s the magic…

👉 Claude saves EVERYTHING.

    Your prompts
    Your code
    Your conversation with AI
    
    Usually these sessions are stored at some config files in Claude , if u want to store these
    sessions into project directory :
    use / export [file_name]
    use /export file.md
    
    Importance of this file :
    You can provide this file as a context to Claude & then u can plan changes in your code..
    

See every session has:

    👉 A unique ID
    👉 Full history



    Now You may ask : why sessions are important?

    Because of something called… 👉 Context Window
    (which you studied this later 👀)

See , its a good practice not to mix/build everything in one session…

👉 otherwise Claude gets confused
👉 and Context gets messy


👉 So its a good idea to create One session = One task


[Show Udemy Course Content img]
Imagine u are building a Learning Platform or a Scalable product like Udemy :
    so u can create one session each in Claude for every feature of Udemy : 
    like :
    
    One session for this feature : Learning Tools
    One session for this feature : Notes
    One seesion for Announcements feature 
    
    
    So always create 1 session each for 1 featue , once it is build , close that session !
    
    
    
    This keeps everything clean..

    This is how real AI engineers work.


'''
```

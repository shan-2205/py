# Lecture 3 : Writing CLAUDE.md — The Brain of Our Agent

```python
'''

"Okay alright, so here is a question for you.

If you hired a new analyst at your company…
What is the FIRST thing you would do?

You would tell them:
    'Here is your job description.'
    'Here are the rules of the company.'
    'Here is how I want our reports formatted.'

Right?

CLAUDE.md is EXACTLY that.

It is a simple text file that you place in your project folder.
And every time Claude Code starts in that folder…
👉 It reads this file FIRST.

So Claude knows:
    Who it is.
    What its job is.
    How it should behave.

This is one of the most POWERFUL features of Claude Code.
And most beginners never use it.


Let's create it now.

In VS Code, in your project root folder — not inside data, not inside reports —
create a new file called:   CLAUDE.md

Capital letters — CLAUDE.md


Now type this inside it:

-------------------------------------------

# Data Analyst Agent : (it means we are assigning Claude a Data Analyst ROLE)

You are an expert data analyst with 10 years of experience.
Your job is to analyze CSV files and produce clear, professional reports.

## When analyzing a CSV file, always follow these steps: ( so here we have to design the wordflow)

1. Read the CSV file from the data/ folder using bash
2. Profile every column — data type, count, null values, unique values, min and max
3. Find data quality issues — duplicates, missing values, incorrect types
4. Write 3 specific chart suggestions based on this data
5. Save the full report to reports/analysis_report.md

## Output format: (this part tells : HOW Claude writes the report.)
- Use clean markdown with headers
- Be specific — show actual numbers, not vague summaries
- Keep language simple so a non-technical person can understand

-------------------------------------------


Now let me explain each part of this file.


The first line says:
    'You are an expert data analyst with 10 years of experience.'

    👉 This is called giving Claude a ROLE.
    👉 When you give Claude a role, it responds like that expert.
    👉 A beginner analyst and an expert analyst give VERY different answers.
    👉 So always give Claude a strong, specific role.


The numbered steps 1 through 5:

    👉 This is the WORKFLOW of our agent.
    👉 We are telling Claude exactly what to do, in what order.
    👉 Claude will follow this sequence every single time.
    👉 No guessing. No missing steps. Always consistent.


The Output format section:

    👉 This controls HOW Claude writes the report.
    👉 Without this, reports can be too long, too technical, or poorly formatted.
    👉 With this, every report looks clean and professional.


Save the file.

Now your project folder looks like this:

    csv_analyst_agent/
    ├── CLAUDE.md           ← the brain of our agent
    ├── data/
    │   └── sales_data.csv
    └── reports/


In the next lecture, we will write the PROMPT.
The single command that triggers our entire agent.

This is where everything comes together."

'''
```

# Lecture 4 : Writing the Agent Prompt (Core Logic)

```python
'''

"Alright — this is the most important lecture of this entire project.

This is where we write the PROMPT.
The single instruction that makes our agent come alive.

And I want to make you understand something before we type anything.

Most people think prompting means just typing a question.
Like… 'analyze this CSV file.'

That works… but the output is messy, incomplete, and unpredictable.

A REAL prompt — the kind AI Engineers write —
has STRUCTURE.

And today I will teach you that structure using our actual prompt.

So type this prompt:


-------------------------------------------

You are an expert data analyst.
Analyze the CSV file in the data/ folder.

Tasks:
1. Read the file and display the first 5 rows
2. Profile each column — show data type, null count, and unique value count
3. Find and list all data quality issues
4. Suggest 3 visualizations that would be useful for this data
5. Save the complete report to reports/analysis_report.md

-------------------------------------------


Now let me break this down piece by piece.


First piece —
    'You are an expert data analyst.'

    👉 This is the CONTEXT.
    👉 We are telling Claude which ROLE to play.
    👉 This is exactly what we did in Project 1 — 'You are an expert Python debugger.'
    👉 Role = better, more specific output.
    👉 Always start your prompt with a strong role.


Second piece —
    'Analyze the CSV file in the data/ folder.'

    👉 This is the INSTRUCTION.
    👉 One clear sentence. No confusion.
    👉 We also told Claude exactly WHERE the file is — data/ folder.
    👉 The more specific you are, the better Claude performs.


Third piece —
    The Tasks list — items 1 through 5.

    👉 This is the OUTPUT INDICATOR.
    👉 We are telling Claude exactly what to produce and in what order.
    👉 Task 1: Show me the data. Let me see what we are working with.
    👉 Task 2: Profile each column. This is the core analysis.
    👉 Task 3: Find the problems. This is the quality check.
    👉 Task 4: Chart suggestions. This adds real business value.
    👉 Task 5: Save the report. So we have something to share.

    👉 Without numbered tasks, Claude might skip steps or mix things up.
    👉 With numbered tasks, Claude works through them one by one.
    👉 Predictable. Reliable. Professional.


This 4-part structure is something you should memorize:

    Context    →   Who Claude should be
    Instruction →  What to do
    Input      →   Where the data is
    Output     →   How to respond


Use this for EVERY prompt you ever write.
Not just in this course — in everything.

In the next lecture, we hit Enter and watch the magic happen."

'''
```

# Lecture 6 : Making It Reusable — The /analyze Custom Command

```python
'''

"Alright — this is a BONUS level lecture.

What we have built so far is already powerful.
But there is one small problem.

Every time you get a new CSV file…
You have to type that entire long prompt again.

That is annoying.

What if instead… you could just type:

    /analyze

And Claude does EVERYTHING automatically?


That is called a CUSTOM SLASH COMMAND.

We hav covered Slash Commands in Section 4 of this course.
Now we are going to USE them in a real project.


Here is how to create it:


STEP 1 — Create the .claude folder

    In your project root, create a new folder called:  .claude
    This is a hidden folder that Claude Code reads automatically.

    Inside .claude, create another folder called:   commands

    Your structure now looks like:

        csv_analyst_agent/
        ├── .claude/
        │   └── commands/
        ├── CLAUDE.md
        ├── data/
        └── reports/


STEP 2 — Create the command file

    Inside  .claude/commands/  create a file called:   analyze.md

    Inside this file, paste exactly this:


    -------------------------------------------

    You are an expert data analyst.
    Analyze the CSV file in the data/ folder.

    Tasks:
    1. Read the file and display the first 5 rows
    2. Profile each column — show data type, null count, and unique value count
    3. Find and list all data quality issues
    4. Suggest 3 visualizations that would be useful for this data
    5. Save the complete report to reports/analysis_report.md

    -------------------------------------------


    Recognize this?
    👉 It is the EXACT SAME prompt we wrote in the last lecture.
    👉 We are just saving it as a reusable command now.


STEP 3 — Use the command

    Go back to your Claude terminal.
    Now instead of typing that long prompt…
    Just type:

        /analyze

    And press Enter.

    -- [show screen] --

    Same result!
    Claude runs the entire analysis with just ONE word.


Now here is the beautiful part.

    Next week your boss gives you a DIFFERENT CSV file.
    You just:
        👉 Drop it in the data/ folder
        👉 Type  /analyze
        👉 Get a full report in 2 minutes

    No setup. No rewriting prompts. Just  /analyze  and done.


This is the REAL power of Claude Code.

    You are not just using AI.
    You are BUILDING systems with AI.
    Systems that work for you, every single day.


So Congratulations — you have just built your Another real AI Agent workflow!



See you in Project 3!"

'''
```

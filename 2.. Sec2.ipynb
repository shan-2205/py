# 3.. What is Agentic Coding (VERY simple) ? 

```python
'''

Hi there , 
"till now you understand how Claude Code works…
Let’s talk about the most exciting concept — Agentic Coding."


Let’s break it"Agentic Coding" down :

    👉 Agent = AI that can perform tasks
    👉 Agentic = Multiple agents like agent1 , agent2 , agent3 & they are  working together

That’s it..

    Coding : set of instructions/Code which u give to system or to Agent to perform certain tasks !



Now imagine this…
👉 You want to build a website like Udemy.

    A full platform where:

    Users can sign up
    Instructors can upload courses
    Students can watch videos
    
    normally… , This is a HUGE project.

    You would need:

        Frontend developer
        Backend developer
        Database engineer
        DevOps

        And it could take Week or Months to build a such a Scalable Product like Udemy !

    But now…

    Let’s do the same thing using Agentic Coding.
    Instead of doing everything yourself…

    👉 You create a TEAM of AI agents.
            OR
    You assign tasks by creating multiple agents : 

    👨‍💻 Agent 1 → Build UI
    👨‍💻 Agent 2 → Handle backend
    👨‍💻 Agent 3 → Database Agent




SO Claude can create :
👉 Sub-agents
👉 which work in parallel & build all these things..


& This is amazing !

Because:

👉 Your Work becomes faster
👉 Tasks don’t mix up
👉 System becomes scalable


Now think like a manager…
You don’t have to code everything…

👉 You just assign work & aegnts will do it..


That is : 👉 Agentic Coding


'''
```
